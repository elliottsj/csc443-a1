## Sequential write to file

We experimented writing a 50-mebibyte (MiB) file to three different mediums (solid-state drive, USB flash drive, hard disk drive) using various block sizes ranging from 128 bytes to 2 MiB.

### Solid-state drive
`./experiment_create_random_file.py` was run on a 2014 MacBook Pro using a solid-state drive with results collected in a file `results_ssd.csv`, summarized below:

```python
%matplotlib notebook

import csv
import numpy as np
import matplotlib.pyplot as plt

def get_data(csvfile):
    reader = csv.DictReader(csvfile)
    return zip(*(
        (
            row['block_size'], 
            row['milliseconds_elapsed'], 
            (
                (int(row['total_size']) / 2 ** 20) # convert bytes -> mebibytes
                / 
                (int(row['milliseconds_elapsed']) / 1000) # convert milliseconds -> seconds
            )
        )
        for row in reader
    ))

with open('./results_ssd.csv') as csvfile:
    x, y, rates = get_data(csvfile)
    
    plt.suptitle('Writing a 50MiB file to a solid-state drive', fontsize=20)
    
    plt.subplot(2, 1, 1)
    plt.title('Time to write a 50MiB file using various block sizes')
    plt.xlabel('Block Size (Bytes)')
    plt.ylabel('Milliseconds Elapsed')
    plt.xscale('log')
    plt.tight_layout()
    plt.plot(x, y, 'bo')
    
    plt.subplot(2, 1, 2)
    plt.title('Data rate writing a 50MiB file using various block sizes')
    plt.xlabel('Block Size (Bytes)')
    plt.ylabel('Data Rate (Mebibytes/Second)')
    plt.xscale('log')
    plt.tight_layout()
    plt.plot(x, rates, 'bo')
    
    plt.subplots_adjust(top=0.85)
    plt.show()
```
